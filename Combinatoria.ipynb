```python
#M贸dulo de importaci贸n
import numpy as np 
from math import factorial as fact
from matplotlib import pyplot as plt 
%matplotlib inline
import timeit
```

Consider a box with three balls, each with a different value: ball A is worth 2
points, ball B is worth 3 points and ball C is worth 4 points. Write a function that
calculates the number of different ways of reaching a total sum of N points by
extracting from the box one ball at a time, taking into account the order in which the
balls are drawn.
Your function must receive the target number of points N as a parameter and
return the number of different ways of getting to that number. Make the function as
efficient as possible and discuss its computational complexity.

```python
#Target number
def ballways(N):

    ACUM_PERMUTATIONS=[]
#extracting the collection of balls that satisfies the target
    for i in range(N+1):
        B=i
        for j in range(N+1):
            C=j
            #auxiliar variable
            TAU=N-3*B-4*C
       
            if TAU % 2 == 0 and TAU>=0:
                A=int(TAU/2)
                SUMA=sum([A,B,C])
                PERMUTATIONS=int(fact(SUMA)/(fact(A)*fact(B)*fact(C)))
                ACUM_PERMUTATIONS=np.append(ACUM_PERMUTATIONS,PERMUTATIONS)
            
    NUM_OF_WAYS=int(sum(ACUM_PERMUTATIONS))

    return NUM_OF_WAYS 
           
            
            
     
            

        
    
```

```python
Acum_time=[]
Acum_n=[]
for j in range(500):
    N=j
    
    start = timeit.default_timer()
    ballways(N)
    end=timeit.default_timer()
    t=end-start
    Acum_n=np.append(Acum_n,N)
    Acum_time=np.append(Acum_time,t)
    
```

```python
plt.scatter(Acum_n,Acum_time)
plt.xlabel('Param N')
plt.ylabel('Execution time (s)')
plt.title('Comportamiento de la funci贸n ballways')
plt.grid()
```

```python
t_ij=Acum_time/Acum_n
```
